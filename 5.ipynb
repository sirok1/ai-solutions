```python
import math

class TrigonometryCalculator:
    def __init__(self):
        pass

    def calculate_cosine(self, angle):
        return math.cos(math.radians(angle))

    def calculate_sine(self, angle):
        return math.sin(math.radians(angle))

    def calculate_tangent(self, angle):
        return math.tan(math.radians(angle))

    def calculate_arcsine(self, value):
        return math.degrees(math.asin(value))

    def calculate_arccosine(self, value):
        return math.degrees(math.acos(value))

    def calculate_arctangent(self, value):
        return math.degrees(math.atan(value))

    def convert_degrees_to_radians(self, degrees):
        return math.radians(degrees)
    
def interactive_trigonometry_calculator():
    calculator = TrigonometryCalculator()

    print("Выберите функцию:")
    print("1. Косинус")
    print("2. Синус")
    print("3. Тангенс")
    print("4. Арксинус")
    print("5. Арккосинус")
    print("6. Арктангенс")
    print("7. Перевод из градусов в радианы")
    
    choice = input("Введите номер функции (1-7): 1. Косинус, 2. Синус, 3. Тангенс, 4. Арксинус, 5. Арккосинус, 6. Арктангенс, 7. Перевод из градусов в радианы")

    if choice in {'1', '2', '3', '4', '5', '6'}:
        

        if choice == '1':
            angle = float(input("Введите угол в градусах: "))
            result = calculator.calculate_cosine(angle)
            print(f"Косинус угла {angle} градусов: {result}")
        elif choice == '2':
            angle = float(input("Введите угол в градусах: "))
            result = calculator.calculate_sine(angle)
            print(f"Синус угла {angle} градусов: {result}")
        elif choice == '3':
            angle = float(input("Введите угол в градусах: "))
            result = calculator.calculate_tangent(angle)
            print(f"Тангенс угла {angle} градусов: {result}")
        elif choice == '4':
            angle = float(input("Введите число от -1 до 1: "))
            result = calculator.calculate_arcsine(angle)
            print(f"Арксинус значения {angle}: {result} градусов")
        elif choice == '5':
            angle = float(input("Введите число от -1 до 1: "))
            result = calculator.calculate_arccosine(angle)
            print(f"Арккосинус значения {angle}: {result} градусов")
        elif choice == '6':
            angle = float(input("Введите число от -1 до 1: "))
            result = calculator.calculate_arctangent(angle)
            print(f"Арктангенс значения {angle}: {result} градусов")
    elif choice == '7':
        degrees = float(input("Введите значение в градусах для перевода в радианы: "))
        result = calculator.convert_degrees_to_radians(degrees)
        print(f"{degrees} градусов равны {result} радианам")
    else:
        print("Неверный выбор функции.")

interactive_trigonometry_calculator()
```

```python
class Tree:
    def __init__(self, kids, next=None):
        self.kids = self.val = kids
        self.next = next
        
t = Tree(Tree('a', Tree('b', Tree('d', 'e'))), Tree('c', Tree('f')))

print('root = ', t.kids.val)

print('left subtree = ', end='')
print(t.kids.next.val, end=', ')
print(t.kids.next.next.val, end=', ')
print(t.kids.next.next.next)

print('right subtree = ', end='')
print(t.next.kids, end=', ')
print(t.next.next.kids)
```

### Задание

```python
class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
    
    def PrintTree(self):
        if self.left:
            self.left.PrintTree()
        print(self.data)
        if self.right:
            self.right.PrintTree()
    
    def AddElement(self, Element):
        if self.data:
            if Element < self.data:
                if self.left is None:
                    self.left = Tree(Element)
                else:
                    self.left.AddElement(Element)
            elif Element > self.data:
                if self.right is None:
                    self.right = Tree(Element)
                else:
                    self.right.AddElement(Element)
        else:
            self.data = Element
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier, plot_tree
import numpy as np

# данные
X_test = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
y_train = [0, 0, 0, 1, 1, 1]

X_test.shape

classifier = DecisionTreeClassifier()
classifier.fit(X_test,y_train)

plot_tree(classifier)

y_pred = classifier.predict(X_test)
y_pred

print(confusion_matrix(y_pred,y_train))
print(classification_report(y_pred,y_train))
```

```python
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import pandas as pd

url = 'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv'
dataframe = pd.read_csv(url)
dataframe.head()
dataframe.describe()

# Нарисуем точечную диаграмму
dataframe.plot(x="Petrol_tax",y=['Paved_Highways','Petrol_Consumption','Population_Driver_licence(%)'],kind="bar")
plt.xlabel("Штаты")
plt.ylabel("Потребление бензина")
plt.show()

X = dataframe[['Average_income','Population_Driver_licence(%)','Petrol_Consumption']]
y = dataframe['Petrol_tax']
print(X)
print(y)

# Теперь, когда у нас есть атрибуты и метки, необходимо разделить их на а обучающий и тестовый наборы.
# Приведенный фрагмент разделяет 80% данных на обучающий набор, а 20% данных - на набор тестов
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# далее можно обучить алгоритм линейной регрессии
# необходимо импортировать класс LinearRegression, создать его экземпляр и вызвать метод fit()
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)

from sklearn import tree
tree.plot_tree(regressor)

#Построим прогноз:
y_pred = regressor.predict(X_test)
y_pred

#Теперь сравним некоторые из наших прогнозируемых значений с фактическими значениями:
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df

#Расчитаем среднюю абсолютную и среднеквадратичную ошибку регрессии:
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))

metrics.mean_absolute_error(y_test, y_pred) / np.average(y) * 100
```
